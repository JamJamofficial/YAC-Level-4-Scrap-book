```python
import tkinter as tk
r = tk.Tk()
r.title("counting seconds")
button = tk.Button(r, text="stop", width=25, command=r.destroy)
button.pack()
r.mainloop()
```

```python
from tkinter import *
master = Tk()
var1 = IntVar()
Checkbutton(master, text='male', variable=var1).gid(row=0,sticky=W)
var2 = IntVar()
Checkbutton(master, text='female', variable=var2).grid(row=1, sticky)
mainloop()
```

```python
from tkinter import *
top = Tk()
Lb = Listbox(top)
Lb.insert(1, 'Python')
Lb.insert(2, 'Java')
Lb.insert(4, 'Any other')
Lb,pack()
top.mainloop()
```
