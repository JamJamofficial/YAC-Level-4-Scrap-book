```python
import numpy as np
import datetime

# Standard Python
start_time = datetime.datetime.now()

rolls = []
for _ in range(1000000000):  # Roll the die a billion times
    roll = np.random.randint(1, 7)  # Generate a random integer between 1 and 6
    rolls.append(roll)

mean = np.mean(rolls)

end_time = datetime.datetime.now()
elapsed_time = end_time - start_time

print(f"Mean (standard Python): {mean}")
print(f"Elapsed time (standard Python): {elapsed_time}")

# NumPy
start_time = datetime.datetime.now()

rolls = np.random.randint(1, 7, size=1000000000)  # Generate a billion random rolls at once
mean = np.mean(rolls)

end_time = datetime.datetime.now()
elapsed_time = end_time - start_time

print(f"Mean (NumPy): {mean}")
print(f"Elapsed time (NumPy): {elapsed_time}")


#ETC (Estimated time of completion) 1h 45min
```

```python
import numpy as np
import datetime

# NumPy
start_time = datetime.datetime.now()

rolls = np.random.randint(1, 7, size=1_000_000_000)  # Generate a billion random rolls at once
mean = np.mean(rolls)

end_time = datetime.datetime.now()
elapsed_time = end_time - start_time

print(f"Mean (NumPy): {mean}")
print(f"Elapsed time (NumPy): {elapsed_time}")
```
