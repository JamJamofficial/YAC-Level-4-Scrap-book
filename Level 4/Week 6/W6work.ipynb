Intro to Matplotlib
legend = key

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([1,2,3,4,5])

x = np.linspace(0, 10, 100)

y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.legend()
plt.show()
```
