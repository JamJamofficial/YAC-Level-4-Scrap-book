```python
import tkinter as tk
import os

# Create the main window
root = tk.Tk()
root.title("Tkinter Input")

# Function to handle submit button click and save to text file
def submit():
    first = first_name_entry.get().strip()
    middle = middle_name_entry.get().strip()
    last = last_name_entry.get().strip()
    age = age_entry.get().strip()

    full_name = " ".join([n for n in (first, middle, last) if n])
    if not full_name:
        result_label.config(text="Please enter at least one name field.")
        return

    result_label.config(text=f"Hello {full_name}! You are {age} years old.")

    try:
        save_path = os.path.join(os.getcwd(), "saved_inputs.csv")
        with open(save_path, "a", encoding="utf-8") as f:
            f.write(f"Name: {full_name}, Age: {age}\n")
    except Exception as e:
        result_label.config(text=f"Error saving: {e}")

label_fn = tk.Label(root, text="Enter your first name:",font = ("Papyrus",30))
label_fn.pack(pady=2)
first_name_entry = tk.Entry(root)
first_name_entry.pack(pady=2)

label_mn = tk.Label(root, text="Enter your middle name:",font = ("Papyrus",30))
label_mn.pack(pady=2)
middle_name_entry = tk.Entry(root)
middle_name_entry.pack(pady=2)

label_ln = tk.Label(root, text="Enter your last name:",font = ("Papyrus",30))
label_ln.pack(pady=2)
last_name_entry = tk.Entry(root)
last_name_entry.pack(pady=2)

label_age = tk.Label(root, text="Enter your age:",font = ("Papyrus",30))
label_age.pack(pady=5)
age_entry = tk.Entry(root)
age_entry.pack(pady=5)

result_label = tk.Label(root, text="")
result_label.pack(pady=5)

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack(pady=10)

# Run the application
root.mainloop()
```
