## Text editor
A text editor that uses the tkinter lib.

```python
import tkinter as tk
from tkinter.filedialog import askopenfilename, asksaveasfilename

def open_file():
    """Open a file."""
    # get the selected filepath
    filepath = askopenfilename(
        filetypes=[("Text Files", "*.txt"), ("All Files", "*.*")]
    )
    if not filepath:
        return  # User cancelled the dialog
    # Clear existing text and insert 
    txt_edit.delete(1.0, tk.END)
    with open(filepath, "r") as input_file:
        text = input_file.read()
        txt_edit.insert(tk.END, text)
    # Update window title 
    window.title(f"Text Editor - {filepath}")

def save_file():
    """Save the contents of the text widget to a file (Save As...)."""
    # Show save-as dialog 
    filepath = asksaveasfilename(
        defaultextension="txt",
        filetypes=[("Text Files", "*.txt"), ("All Files", "*.*")],
    )
    if not filepath:
        return  # User cancelled the dialog
    # Write current text widget contents 
    with open(filepath, "w") as output_file:
        text = txt_edit.get(1.0, tk.END)
        output_file.write(text)
    # Update window title to show saved file path
    window.title(f"Text Editor - {filepath}")

# Create main application window
window = tk.Tk()
window.title("Text Editor")
# Configure row/column weights so the text widget expands with the window
window.rowconfigure(0, minsize=800, weight=1)
window.columnconfigure(1, minsize=800, weight=1)

# Main text editing widget
txt_edit = tk.Text(window)

# Frame to hold buttons
fr_buttons = tk.Frame(window, relief=tk.RAISED, bd=2)
btn_open = tk.Button(fr_buttons, text="Open", command=open_file)
btn_save = tk.Button(fr_buttons, text="Save As...", command=save_file)

# Layout buttons inside the frame using grid
btn_open.grid(row=0, column=0, sticky="ew", padx=5, pady=5)
btn_save.grid(row=1, column=0, sticky="ew", padx=5)

# Place the frame and text widget in the main window using grid
fr_buttons.grid(row=0, column=0, sticky="ns")
txt_edit.grid(row=0, column=1, sticky="nsew")

# Start the application loop
window.mainloop()
```
