## Create an image of 200 by 200 and draw a square with white colour and an area of 250 squared unit using NumPy and matplotlib.  Visualise your work and submit the jupyter notebook

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 200x200 image
image = np.zeros((200, 200, 3), dtype=np.uint8)

# Define the square parameters
square_size = int(np.sqrt(250))  # Calculate the side length 

# Calculate the starting point 
start_x = (200 - square_size) // 2
start_y = (200 - square_size) // 2

# Draw the square with white colour
image[start_y:start_y + square_size, start_x:start_x + square_size] = [255, 255, 255]

# plot the image
plt.imshow(image)
```

## Simulate ludo by demonstrate theefficient perforance of numpy if the dice is tossed 100 million times.

```python
# define the amount of tosses
def simulate_ludo(tosses=100_000_000):
    # Simulate dice rolls
    rolls = np.random.randint(1, 7, size=tosses)
    
    # Count how many times each face appears
    counts = np.bincount(rolls)[1:]  
    
    return counts
```

## Simulate what the total score will be if a dice was tosed 9 billion times. 

```python
def simulate_total_score(tosses=9_000_000_000):
    # Simulate dice rolls
    rolls = np.random.randint(1, 7, size=tosses)
    
    # Calculate total score
    total_score = np.sum(rolls)
    
    return total_score
```

## What is serialisation in fastapi? Give example.

- serialisation in fastapi is converting complex data types into a format that can be esily shared and stored

  example:

  from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

# Define a Pydantic model for the data
class Item(BaseModel):
    name: str
    description: str | None = None
    price: float
    tax: float | None = None

# Create a sample data
items = {
    "foo": {"name": "Foo", "description": "The pretender", "price": 50.2, "tax": 10.2},
    "bar": {"name": "Bar", "description": "The protector", "price": 60.2, "tax": 12.2},
    "baz": {"name": "Baz", "description": "The conqueror", "price": 70.2, "tax": 14.2},
}

@app.get("/items/{item_id}", response_model=Item)
async def read_item(item_id: str):
    """
    This endpoint retrieves an item by its ID.
    It uses the `Item` model for serialization.
    """
    if item_id in items:
        return items[item_id]
    else:
        return {"message": "Item not found"}
